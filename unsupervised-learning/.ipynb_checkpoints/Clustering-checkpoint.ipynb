```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from tqdm.notebook import tqdm
from matplotlib import pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.metrics.cluster import homogeneity_score
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px
from sklearn.decomposition import PCA
from sklearn.decomposition import FastICA
import scipy
```

# Clustering 

## Loading Datasets

```python
df_mocap = pd.read_csv("data/Postures.csv").sample(10000, random_state=27)
df_mocap.replace("?", 0, inplace=True)


X_mocap = StandardScaler().fit_transform(df_mocap.drop(['Class','User'], axis=1))
y_mocap = df_mocap["Class"]

# X_train_pos, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1101)
```

```python
df_pd = pd.read_csv("data/pd_speech_features.csv")

X_pd = StandardScaler().fit_transform(df_pd.drop(['class','id'], axis=1))
y_pd = df_pd["class"]
```

```python
#reference: https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html
def get_kmeans_elbow(X, title):
    clusters = [x for x in range(2,10)]
    sil_scores = {}
    for cluster in tqdm(clusters):
        clusterer = KMeans(n_clusters=cluster, random_state=27)
        cluster_labels = clusterer.fit_predict(X)
        sil_score = silhouette_score(X, cluster_labels)
        sil_scores[cluster] = sil_score
    plt.plot(sil_scores.keys(), sil_scores.values(), label="kmeans")
    
    sil_scores = {}
    for cluster in tqdm(clusters):
        clusterer = GaussianMixture(n_components=cluster, random_state=27, reg_covar=1e-4)
        cluster_labels = clusterer.fit_predict(X)
        sil_score = silhouette_score(X, cluster_labels)
        sil_scores[cluster] = sil_score
    plt.plot(sil_scores.keys(), sil_scores.values(), label="gmm")
    plt.xlabel("No. of clusters")
    plt.ylabel("Silhoutte Coefficient")
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
get_kmeans_elbow(X_pd, "Silhoutte Coefficient vs K for PD dataset")
```

```python
get_kmeans_elbow(X_mocap, "Silhoutte Coefficient vs K for Mocap dataset")
```

# DimRed

## PCA

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (8,6)

def get_pca(X, limit, title):
    exp = {}
    for i in range(1, limit+1):
        pca = PCA(n_components=i)
        pca.fit(X)
        exp[i] = sum(pca.explained_variance_ratio_)
    plt.plot(exp.keys(), exp.values())
    plt.xlabel("No. of components")
    plt.ylabel("Explained Variance Ratio")
    plt.title(title)
    plt.show()
```

```python
get_pca(X_mocap, 36, "Total EVR vs Number of components (MoCap)")
```

```python
pca = PCA(n_components=20)
pca.fit(X_mocap)

df_weights = pd.DataFrame(pca.components_,columns=df_mocap.drop(['Class','User', 'cluster'], axis=1).columns)
df_weights.describe()
```

```python
get_pca(X_pd, 200, "Total EVR vs Number of components (PD)")
```

## ICA

```python
import numpy as np
def get_ica(X, limit, title):
    exp = {}
    for i in range(1, limit+1):
        pca = FastICA(n_components=i, random_state=27).fit_transform(X)
        exp[i] = np.mean(scipy.stats.kurtosis(pca))
    plt.plot(exp.keys(), exp.values())
    plt.xlabel("No. of components")
    plt.ylabel("Average Kurtosis")
    plt.title(title)
    plt.show()
```

```python
get_ica(X_mocap, 36, "Avg. Kurtosis vs Number of components (MoCap)")
get_ica(X_pd, 200, "Avg. Kurtosis vs Number of components (PD)")
```

```python
pca = FastICA(n_components=25, random_state=27)
pca.fit(X_mocap)

df_weights = pd.DataFrame(pca.components_,columns=df_mocap.drop(['Class','User', 'cluster'], axis=1).columns)
df_weights.describe()
```

## RCA

```python
from sklearn.random_projection import GaussianRandomProjection
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
def get_rca(X, limit, title):
    exp = {}
    for i in range(1, limit+1):
        rca = GaussianRandomProjection(n_components=i, random_state=44)
        x_rca = rca.fit_transform(X)

        reconstructed = np.dot(x_rca, np.linalg.pinv(rca.components_.T))
        rc_err = mean_squared_error(X, reconstructed)    
        exp[i] = rc_err   

    plt.plot(exp.keys(), exp.values())
    plt.xlabel("No. of components")
    plt.ylabel("Reconstruction Error")
    plt.title(title)
    plt.show()
```

```python
get_rca(X_mocap, 36, "Reconstruction Error vs Number of components (MoCap)")
get_rca(X_pd, 200, "Reconstruction Error vs Number of components (PD)")
```

```python
pca = GaussianRandomProjection(n_components=20, random_state=44)
pca.fit(X_mocap)

df_weights = pd.DataFrame(pca.components_,columns=df_mocap.drop(['Class','User', 'cluster'], axis=1).columns)
df_weights.describe()
```

## Applying DimRed to Clustering

```python
def compare_dimreds(X, exp_name):
    clusters = [x for x in range(2,10)]
    sil_scores = {}
    for cluster in tqdm(clusters):
        clusterer = KMeans(n_clusters=cluster, random_state=27)
        cluster_labels = clusterer.fit_predict(X)
        sil_score = silhouette_score(X, cluster_labels)
        sil_scores[cluster] = sil_score
    plt.plot(sil_scores.keys(), sil_scores.values(), label="{}-kmeans".format(exp_name))
    
    sil_scores = {}
    for cluster in tqdm(clusters):
        clusterer = GaussianMixture(n_components=cluster, random_state=27, reg_covar=1e-4)
        cluster_labels = clusterer.fit_predict(X)
        sil_score = silhouette_score(X, cluster_labels)
        sil_scores[cluster] = sil_score
    plt.plot(sil_scores.keys(), sil_scores.values(), label="{}-gmm".format(exp_name))
```

### Comparing on PD Dataset

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,8)

dataset = X_pd
compare_dimreds(dataset, "Baseline")
compare_dimreds(PCA(n_components=100).fit_transform(dataset), "PCA")
compare_dimreds(FastICA(n_components=75, random_state=27).fit_transform(dataset), "ICA")
compare_dimreds(GaussianRandomProjection(n_components=200, random_state=44).fit_transform(dataset), "RCA")
compare_dimreds(df_pd[pd_list], "DT")
plt.xlabel("No. of clusters")
plt.ylabel("Silhoutte Coefficient")
plt.title("Silhoutte Coefficient vs K for PD dataset")
plt.legend(loc="best")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,8)

dataset = X_mocap
compare_dimreds(dataset, "Baseline")
compare_dimreds(PCA(n_components=20).fit_transform(dataset), "PCA")
compare_dimreds(FastICA(n_components=25, random_state=27).fit_transform(dataset), "ICA")
compare_dimreds(GaussianRandomProjection(n_components=25, random_state=44).fit_transform(dataset), "RCA")
compare_dimreds(df_mocap[mocap_list], "DT")
plt.xlabel("No. of clusters")
plt.ylabel("Silhoutte Coefficient")
plt.title("Silhoutte Coefficient vs K for MoCap dataset")
plt.legend(loc="best")
plt.show()
```
